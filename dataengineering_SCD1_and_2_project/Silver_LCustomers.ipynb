```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
df = spark.read.format('parquet').load('abfss://bronze@datalakrelatestreddy.dfs.core.windows.net/customers')
```

```python
display(df)
```

```python
df = df.drop("_rescued_data")
```

```python
df = df.withColumn('full_name', concat(col('first_name'), lit(' '), col('last_name')))
```

```python
df.display()
```

```python
df = df.drop("first_name", "last_name")
```

```python
df.display()
```

```python
df = df.withColumn('domain',split(col("email"), '@')[1])
df.display()
```

```python
%sql
create schema if not exists vishnudatabrickslatest.silver
```

```python
df.write.format('delta').mode('overwrite').option('path','abfss://silver@datalakrelatestreddy.dfs.core.windows.net/customers').saveAsTable('vishnudatabrickslatest.silver.customers')
```
