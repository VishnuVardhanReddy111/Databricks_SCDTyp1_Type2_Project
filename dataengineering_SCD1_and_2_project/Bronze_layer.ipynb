```python
df = spark.readStream.format('cloudFiles').option('cloudFiles.format','parquet')\
                    .option('cloudFiles.schemaLocation','abfss://bronze@datalakrelatestreddy.dfs.core.windows.net/schema')\
                    .load('abfss://source@datalakrelatestreddy.dfs.core.windows.net/customers')
```

```python
df.writeStream.format('parquet').outputMode('append')\
    .option('checkpointLocation','abfss://bronze@datalakrelatestreddy.dfs.core.windows.net/checkpoint/customers')\
    .option('path','abfss://bronze@datalakrelatestreddy.dfs.core.windows.net/customers')\
    .trigger(once=True).start()
```
