```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
from pyspark.sql.window import Window
```

```python
df = spark.sql('select * from databricksreddy.bronze.customers')
```

```python
df.display()
```

```python
df = df.drop('_rescued_data')
df.display()
```

```python
df = df.withColumn("full_name",concat(col('first_name'),lit(' '),col("last_name")))
```

```python
df = df.drop("first_name","last_name")
```

```python
df.display()
```

```python
df = df.withColumn('domains',split(col("email"),'@')[1])
```

```python
df.display()
```

```python
df.groupBy("domains").agg(count("domains").alias('domain_count')).sort('domain_count',ascending=False).display()
```

```python
df_gmail = df.filter(col("domains") == 'gmail.com')
df_gmail.display()

df_yahoo = df.filter(col("domains") == 'yahoo.com')
df_yahoo.display()

df_hotmail = df.filter(col("domains") == 'hotmail.com')
df_hotmail.display()
```
