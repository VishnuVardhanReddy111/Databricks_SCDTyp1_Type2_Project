```python
%sql
create schema if not exists databricksreddy.bronze
```

```python
df = spark.read.format('parquet').option('inferSchema','true').load(f'abfss://source@datalakevishnu1.dfs.core.windows.net/{file_name}')
```

```python
df.display()
```

```python
df = spark.readStream.format('cloudFiles').option('cloudFiles.format','parquet').option('cloudFiles.schemaLocation',f'abfss://bronze@datalakevishnu1.dfs.core.windows.net/schema/{file_name}').load(f'abfss://source@datalakevishnu1.dfs.core.windows.net/{file_name}')
```

```python
df.writeStream.format('parquet').outputMode('append').option('checkpointLocation',f'abfss://bronze@datalakevishnu1.dfs.core.windows.net/checkpoints/{file_name}').option('path',f'abfss://bronze@datalakevishnu1.dfs.core.windows.net/{file_name}').trigger(once=True).start()
```
